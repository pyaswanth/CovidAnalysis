```python
import plotly.express as px
import plotly.graph_objects as go
import plotly.figure_factory as ff
from plotly.subplots import make_subplots
import math
import random
from datetime import timedelta
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
df=pd.read_csv("covid_19_india.csv")
df
```

```python
gp_by_date=df.groupby(by=['Date']).sum().reset_index()
gp_by_date.head()
```

```python
gp_by_state=df.groupby(by=['State/UnionTerritory']).sum().reset_index()
gp_by_state
```

```python
import matplotlib.pyplot as plt
x=gp_by_state['State/UnionTerritory']
y=gp_by_state['Confirmed']
plt.plot(x,y)
plt.show()
```

```python
plt.bar(x,y)
plt.show()
```

```python
state_data=df[df['State/UnionTerritory']=='Andhra Pradesh']
per_day_confirmed_cases=state_data['Confirmed'].diff()
state_data['PerDayConfirmed']=per_day_confirmed_cases
state_data
```

```python
per_day_confirmed_cases=state_data['Confirmed'].diff()
state_data['PerDayConfirmed']=per_day_confirmed_cases
x=state_data['Date']
y=state_data['PerDayConfirmed']
fig = go.Figure()
fig.add_trace(go.Scatter(x=x,y=y))
fig.update_layout(title='Covid cases in Andhra',xaxis_tickfont_size=14,yaxis=dict(title='Number of Cases'))
fig.show()
```
